```python
#8
def getContent(soup):
    
    content_text = ''
    content = soup.find("div", {"class":"element element-textarea first"}).findAll('p')
    for element in content:
        content_text += '\n' + ''.join(element.findAll(text = True))
    return content_text

print(getContent(soup))
```

```python
#9
def getBackground(soup):

    background_text = ''
    backgroundLinks = [div for div in 
        soup.findAll('div', attrs={'class' : 'element element-textarea'})]
    for element in backgroundLinks:
        #background_text += '\n' + ''.join(element.findAll(text = True))
         background_text += '\n' + ''.join(element.h3.getText() + ' ' + element.p.getText())
    return background_text

def getConclusion(soup):

    background_text = ''
    backgroundLinks = [div for div in 
        soup.findAll('div', attrs={'class' : 'element element-textarea last'})]
    for element in backgroundLinks:
        #background_text += '\n' + ''.join(element.findAll(text = True))
         background_text += '\n' + ''.join(element.h3.getText() + ' ' + element.p.getText())
    return background_text


print(getBackground(soup) + getConclusion(soup))


```
